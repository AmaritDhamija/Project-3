# Introduction

This project analyzes crime data from Winnipeg using a real-world dataset from an open data source.

The dataset contains information about:
- the year and month of incidents
- neighbourhood and community locations
- crime categories and types
- number of occurrences (count_stats)

I chose this dataset because crime trends are important for understanding how different areas are affected over time. This analysis can help identify patterns, high-risk areas, and trends in different types of crime.

The goal of this project is to:
- explore and clean the dataset
- identify patterns in crime across time and locations
- visualize key insights
- build a simple predictive model

# Data Cleaning

## Missing Values

The dataset contains missing values in:
- Neighbourhood
- Community

Since these columns are important for location-based analysis, I decided to **drop rows with missing values** instead of filling them. Filling them could introduce incorrect location data.

## Column Formatting

Column names were cleaned by:
- converting to lowercase
- replacing spaces with underscores

This improves consistency and avoids errors when referencing columns in code.
